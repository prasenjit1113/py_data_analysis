# SEABORN  (better to visualize data chart)

# Line plot Seaborn

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
data = {"Days":[1,2,3,4,5],
       "NOP":[30,40,20,35,53]}
df = pd.DataFrame(data)
print(df)

sns.lineplot(data = data, x = "Days", y = "NOP")
plt.show()
```

```python
data = pd.read_excel("ESD.xlsx")
# print(data)

color = sns.color_palette("viridis")  #like scatter plot
sns.lineplot(data = data, x = "Business Unit", y="Age", hue = "Ethnicity" ,palette = color)   #hue works as legend in matplotlib
plt.show()
```

# Bar Plot Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
data = sns.load_dataset("tips")  #tips,titanic, etc. are the default datasets of seaborn
print(data)

# sns.barplot(data=data, x='day', y='tip', estimator="median")   #estimator="median","mean","sum" etc.
sns.barplot(data=data, x='day', y='tip', hue="sex", palette="spring", order=["Sun","Sat","Fri","Thur"])  #order-> changes the order
plt.show()
```

# Historgram Plot Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
data = sns.load_dataset("tips")  
# print(data)

sns.histplot(data=data, x='day',hue="sex", kde=True)  #kde used for analyze the plot using lines
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
data = sns.load_dataset("titanic")  
# print(data)
# sns.histplot(data=data, x="age", hue="who", kde="True", discrete=True)  #discrete is used to describe each row in plot
sns.histplot(data=data, x="age", hue="who", kde="True", bins=10)  #bins used to show number of bars we want
plt.show()
```

# Scatter Plot Seaborn  (to implement number to number plotting)

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
data = sns.load_dataset("tips")
print(data)

sns.scatterplot(data=data, x="total_bill", y="tip", hue="time", size="size", marker="d")  #size used to show number of customers come together
plt.show()
```

```python
data = pd.read_excel("ESD.xlsx")
# print(data)

sns.scatterplot(data=data, x="Age", y="Annual Salary", size="Bonus %", hue="Gender")
plt.legend(bbox_to_anchor = (0.3,0,1.2,1.2))  #(x,y,width,height)
plt.show()
```

# Heatmap Plot Seaborn  (it is also used to implement numerical(number to number) plotting)

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
data = sns.load_dataset("tips")
print(data)

#here we cannot directly use x=total_bill ,y="tip"  , instead we have to group by them
gp = data.groupby("day").agg({"tip":"mean"})
sns.heatmap(gp)
plt.show()
```

```python
data = pd.read_excel("ESD.xlsx")
gp = data.groupby("Job Title").agg({"Annual Salary":"mean"})
sns.heatmap(gp)
plt.show()
```

# Count Plot Seaborn  (majorly used for counting, in this plot only x-axis is used)

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
data = sns.load_dataset("tips")
print(data)

sns.countplot(data=data, x="day")
plt.show()
```

```python
df = pd.read_excel("ESD.xlsx")
# print(df)
sns.countplot(data=df, x="Department", hue="Gender", palette="GnBu")
plt.show()
```

# Violin Plot Seaborn  (to analyze statistical data (Mean,Median,Mode))  (used in numerical values)

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
data = sns.load_dataset("tips")
print(data)

sns.violinplot(data=data, x="total_bill")
plt.show()
```

```python
df = pd.read_excel("ESD.xlsx")
sns.violinplot(data=df, x="Annual Salary")
plt.show()
```
