# MATPLOTLIB for Data Visualization

# Bar Plot

```python
import matplotlib.pyplot as plt

y = [98,66,88,55,77]
x = ["Part1","Part2","Part3","Part4","Part5"]

plt.bar(x,y)
plt.xlabel("subjects in MCA")
plt.ylabel("numbers in subjects")  
plt.title("MCA sem1 record")
plt.show()
```

```python
y = [98,66,88,55,77]
x = ["Part1","Part2","Part3","Part4","Part5"]

colors = ["yellow","red","green","blue","pink"]

plt.bar(x,y, color=colors , edgecolor="red")    #color="red"  ,  edgecolor="black"
plt.xlabel("subjects in MCA" , fontsize = 10)
plt.ylabel("numbers in subjects" , fontsize = 15)  #fontsize is used to change font size
plt.title("MCA sem1 record" , fontsize = 20)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_excel("E:/DSA_ML/python_ml/expense3.xlsx")

#pandas group by for grouping the columns we want to show
df = pd.DataFrame(data)
print(df)
grouped_by = df.groupby("Payment Mode")["Amount"].sum()
print(grouped_by)

#plotting
plt.bar(grouped_by.index,grouped_by.values)
plt.show()
```

# Line Plot

```python
import matplotlib.pyplot as plt

x = ["Day1","Day2","Day3","Day4","Day5"]
y = [200,250,150,172,300]
y1 = [300,180,290,270,350]

plt.plot(x,y, marker="o", ls="--", color="red", label = "week1")   #marker="o","d","^"   #label is used for legend
plt.plot(x,y1, marker="*", ls="-", color="green", label="week2" , alpha=0.3)   #alpha is used to decrement the highlight  
plt.legend()
# plt.legend(loc=10)  #to change the location of legend. By default it is 0, the values can be between 0-10
plt.show()
```

# Scatter Plot   (This basically used between numerical values)

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(1,10,50)     #to generate random 50 values from 1 to 10 
y = np.random.randint(10,100,50)   #to generate random 50 values from 10 to 100
color = np.random.randint(10,100,50)  #cmap works when color is provided
size = np.random.randint(10,100,50)   #to define sizes of the markers

plt.scatter(x,y,marker = "*", cmap="cool", c=color , s=size)  
plt.colorbar()    #cmap="cool"/"hot"/"ocean"/"viridis" etc.
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data = pd.read_excel("ESD.xlsx")
df = pd.DataFrame(data)
# print(df)
plt.scatter(df["Annual Salary"], df["EEID"], marker = "*", s = df["Age"])   #defines size on the basis of age
plt.show()
```

# Pie chart  (percentage , Values)

```python
import matplotlib.pyplot as plt

brands = ["Oneplus","Nokia","Redmi","Nothing","Vivo","Apple"]
x = [15,10,50,28,45,25]
c = ["orange","green","yellow","blue","red","violet"]
ex = [0,0,0,0,0,0.1]  #to highlight
plt.pie(x,labels = brands,colors = c,explode = ex,shadow = True, autopct="%.2f", startangle=180)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_excel("E:/DSA_ML/python_ml/expense3.xlsx")
df = pd.DataFrame(data)
grouped_by = df.groupby("Payment Mode")["Amount"].sum()
print(grouped_by)

plt.pie(grouped_by.values,labels = grouped_by.index, autopct="%.2f")
plt.show()
print(df)
```

# Stack Plot

```python
import matplotlib.pyplot as plt
days = [1,2,3,4,5,6,7]
custo1 = [10,15,25,22,28,18,35]
custo2 = [15,20,17,25,22,30,43]
custo3 = [22,27,33,20,35,45,52]

plt.stackplot(days, custo1, custo2, custo3, colors=["blue","yellow","green"], labels = ["week1","week2","week3"])   #baseline="sym" can be used to change allignments
plt.legend(loc=2)   #loc=2 changes the location of legend
plt.show()
```

```python
import pandas as pd
data = pd.read_csv("E:/DSA_ML/python_ml/student_scores.csv")
df = pd.DataFrame(data)

grouped = df.head(50).groupby("ParentEduc")["MathScore", "ReadingScore", "WritingScore"].mean()
print(grouped)

# Ensure 'ParentEduc' values align with grouped data for plotting
parent_educ = grouped.index
plt.stackplot(parent_educ,grouped["MathScore"],grouped["ReadingScore"],grouped["WritingScore"])
plt.show()
```

# Legends -> to define Values into charts

```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50]
y = [22,34,44,29,37]
y1 = [33,44,27,39,52]

plt.figure(figsize=[5,3])   #to increase/decrease size
plt.plot(x,y, label="male", marker="d")
plt.plot(x,y1, label="female",marker="o")
plt.legend()   #loc=1,2,3,4,5.. to change location of legends
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50]
y = [22,34,44,29,37]
y1 = [33,44,27,39,52]
y2 = [23,34,37,19,22]
y3 = [31,42,21,31,59]

plt.figure(figsize=[4,3])   
plt.plot(x,y, label="male")
plt.plot(x,y1, label="female")
plt.plot(x,y2, label="kids")
plt.plot(x,y3, label="others")
#bbox_to_anchor -> used to use legends out of the chart
plt.legend(bbox_to_anchor = (0.8, 1.2),ncol = 2, labelspacing=3)   #ncol is used to devide legends into columns   #labelspacing for spacing between legends
```

# Subplot  (multiple plots in one chart)

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [34,66,24,73,45]
plt.subplot(2,2,1)   #subplot(row no, column no, chart no)
plt.plot(x,y)

x = [6,7,8,9,10]
y = [54,27,44,33,75]
plt.subplot(2,2,2)
plt.plot(x,y)

x = [11,12,13,14,15]
y = [14,47,24,53,85]
plt.subplot(2,2,3)
plt.plot(x,y)

x = [16,17,18,19,20]
y = [34,57,64,23,35]
plt.subplot(2,2,4)
plt.plot(x,y)
plt.show()
```

```python
x = [1,2,3,4,5]
y = [34,66,24,73,45]
plt.subplot(1,2,1)   #subplot(row no, column no, chart no)
plt.title("Age")
plt.plot(x,y)

x = [6,7,8,9,10]
y = [54,27,44,33,75]
plt.subplot(1,2,2)
plt.title("Salary")
plt.bar(x,y)
plt.suptitle("Employee")  #super title
plt.show()
```
