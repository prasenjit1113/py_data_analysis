# Bar Plot

```python
import matplotlib.pyplot as plt

y = [98,66,88,55,77]
x = ["Part1","Part2","Part3","Part4","Part5"]

plt.bar(x,y)
plt.xlabel("subjects in MCA")
plt.ylabel("numbers in subjects")  
plt.title("MCA sem1 record")
plt.show()
```

```python
y = [98,66,88,55,77]
x = ["Part1","Part2","Part3","Part4","Part5"]

colors = ["yellow","red","green","blue","pink"]

plt.bar(x,y, color=colors , edgecolor="black")    #color="red"  ,  edgecolor="black"
plt.xlabel("subjects in MCA" , fontsize = 10)
plt.ylabel("numbers in subjects" , fontsize = 15)  #fontsize is used to change font size
plt.title("MCA sem1 record" , fontsize = 20)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_excel("E:/DSA_ML/python_ml/expense3.xlsx")

#pandas group by for grouping the columns we want to show
df = pd.DataFrame(data)
print(df)
grouped_by = df.groupby("Payment Mode")["Amount"].sum()
print(grouped_by)

#plotting
plt.bar(grouped_by.index,grouped_by.values)
plt.show()
```

# Line Plot

```python
import matplotlib.pyplot as plt

x = ["Day1","Day2","Day3","Day4","Day5"]
y = [200,250,150,172,300]
y1 = [300,180,290,270,350]

plt.plot(x,y, marker="o", ls="--", color="red", label = "week1")   #marker="o","d","^"   #label is used for legend
plt.plot(x,y1, marker="*", ls="-", color="green", label="week2" , alpha=0.5)   #alpha is used to decrement the highlight  
plt.legend()
# plt.legend(loc=10)  #to change the location of legend. By default it is 0, the values can be between 0-10
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_excel("E:/DSA_ML/python_ml/expense3.xlsx")
df = pd.DataFrame(data)

#pandas group by for grouping the columns we want to show
grouped_by = df.groupby("Payment Mode")["Amount"].sum()
print(grouped_by)
print(df)

# plotting
plt.plot(grouped_by.index,grouped_by.values)
plt.show()
```

# Scatter Plot   (This basically used between numerical values)

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(1,10,50)     #to generate random 50 values from 1 to 10 
y = np.random.randint(10,100,50)   #to generate random 50 values from 10 to 100
color = np.random.randint(10,100,50)  #cmap works when color is provided
size = np.random.randint(10,100,50)   #to define sizes of the markers

plt.scatter(x,y,marker = "*", cmap="cool", c=color , s=size)  
plt.colorbar()    #cmap="cool"/"hot"/"ocean"/"viridis" etc.
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data = pd.read_excel("ESD.xlsx")
df = pd.DataFrame(data)
# print(df)
plt.scatter(df["Annual Salary"], df["EEID"], marker = "*", s = df["Age"])   #defines size on the basis of age
plt.show()
```

# Pie chart  (percentage , Values)

```python
import matplotlib.pyplot as plt

brands = ["Oneplus","Nokia","Redmi","Nothing","Vivo","Apple"]
x = [15,10,50,28,45,25]
c = ["orange","green","yellow","blue","red","violet"]
ex = [0,0,0,0,0,0.1]  #to highlight
plt.pie(x,labels = brands,colors = c,explode = ex,shadow = True, autopct="%.2f", startangle=180)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_excel("E:/DSA_ML/python_ml/expense3.xlsx")
df = pd.DataFrame(data)
grouped_by = df.groupby("Payment Mode")["Amount"].sum()
print(grouped_by)

plt.pie(grouped_by.values,labels = grouped_by.index, autopct="%.2f")
plt.show()
print(df)
```

# Box Plot (maximum,minimum frequency)

```python
import matplotlib.pyplot as plt
l=[25,46,57,55,33,11,66,88,75,25,98,44]
plt.boxplot(l)
plt.show()
```

```python
import matplotlib.pyplot as plt
l = [1,3,4,7,12,2,9,8,24]
l1 = [2,3,4,5,7,6,3,1]
plot_values = [l,l1]
plt.boxplot(plot_values)
plt.show()
```
